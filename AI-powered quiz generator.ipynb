```python
# Imports
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain, SequentialChain
import os

os.environ["GOOGLE_API_KEY"] = ""

print("Setup complete!")
```

```python
# Initialize Model

llm = ChatGoogleGenerativeAI(
    model="gemini-2.0-flash",
    temperature=0.7,
    convert_system_message_to_human=True
)

print("Model initialized!")
```

```python
# Question Prompt Template

question_prompt = PromptTemplate(
    input_variables=["topic"],
    template="""
Generate one beginner-level quiz question about the following topic:

Topic: {topic}

The question should be clear, simple, and suitable for beginners.
"""
)

print("Question prompt created!")
```

```python
# Answer Prompt Template

answer_prompt = PromptTemplate(
    input_variables=["question"],
    template="""
Provide a detailed answer to the following question:

Question: {question}

Include:
- The correct answer
- A clear explanation suitable for beginners
"""
)

print("Answer prompt created!")
```

```python
# Create Individual Chains

question_chain = LLMChain(
    llm=llm,
    prompt=question_prompt,
    output_key="question"
)

answer_chain = LLMChain(
    llm=llm,
    prompt=answer_prompt,
    output_key="answer"
)

print("Chains created!")
```

```python
# Sequential Chain

quiz_chain = SequentialChain(
    chains=[question_chain, answer_chain],
    input_variables=["topic"],
    output_variables=["question", "answer"],
    verbose=True
)

print("Sequential chain ready!")
```

```python
# Run Quiz Generator

topic = input("Enter a topic for your quiz: ")

response = quiz_chain.invoke({"topic": topic})

print("\n--- Quiz Question ---")
print(response["question"])

print("\n--- Answer & Explanation ---")
print(response["answer"])
```
